#0. INTRODUCCION

El presente análisis comparativo tiene como propósito principal minimizar el riesgo de inversión y maximizar el potencial de crecimiento sostenible para el Sr. Juan. Dada la existencia de cuatro plataformas de venta (Tienda 1 a Tienda 4) con diferentes niveles de rendimiento, es imperativo fundamentar la decisión de inversión en datos cuantitativos y cualitativos.

Justificación del análisis

La decisión de establecer una operación de venta se basa en el principio de que no todas las plataformas son iguales, incluso si venden productos similares. Un enfoque basado únicamente en el ingreso bruto (el dinero que entra) ignora costos críticos y la calidad del servicio, factores que impactan directamente la rentabilidad neta y la viabilidad a largo plazo. Por lo tanto, el análisis se justifica por la necesidad de evaluar el balance estratégico de cada tienda.

Métricas clave y su justificación

Para lograr una recomendación robusta, se seleccionaron métricas que abarcan el potencial de mercado, la eficiencia operativa y la calidad del ecosistema:

Ingresos Totales (Métrica Financiera):

Justificación: Mide la liquidez total del mercado y el poder adquisitivo agregado de la clientela. Es el indicador primario del tamaño del mercado disponible.

Calificación promedio de clientes (Métrica de Calidad):

Justificación: Refleja la satisfacción del cliente y la calidad del servicio de la plataforma. Un puntaje alto se correlaciona directamente con una mayor lealtad de marca y menores tasas de devolución, crucial para el crecimiento sostenido del Sr. Juan.

Costo de envío promedio (Métrica de Eficiencia Operacional):

Justificación: Determina la competitividad logística de la plataforma. Un costo más bajo se traduce en un mayor margen de ganancia o la capacidad de ofrecer un precio final más bajo que la competencia, impactando directamente la rentabilidad neta.

Ventas por categoría y producto (Métrica de Demanda Específica):

Justificación: Revela los patrones de consumo y el interés de nicho dentro de cada tienda. Permite al Sr. Juan identificar si su línea de productos se alinea con las categorías de mayor facturación y si existe un mercado para productos especializados.

Al integrar estas métricas, podemos movernos de una simple comparación de "quién factura más" a una decisión informada sobre "quién ofrece el mejor retorno ajustado al riesgo y la calidad".

#1. Análisis de facturación



En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.

```python
#Ingreso total por cada tienda, Se calcula por medio de la suma de los precios de los articulos vendidos
lista_de_tiendas = [tienda, tienda2, tienda3, tienda4]

# Usamos 'enumerate' para obtener el índice y el DataFrame
for i, df_tienda in enumerate(lista_de_tiendas, start=1):

    # Usamos el índice 'i' para identificar la tienda en la impresión
    nombre_tienda = print(f'Tienda {i}')
    Ingreso_total= df_tienda.Precio.sum()
    print(f'El ingreso total fue: ${Ingreso_total:,.2f}')
    print("-" * 40)
```

# 2. Ventas por categoría

En este debes calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
#Debo contar el numero de ventas por categoria
lista_de_tiendas = [tienda, tienda2, tienda3, tienda4]
for i, df_tienda in enumerate(lista_de_tiendas, start=1):
  print(f"Tienda {i} - Ventas por categoría")
  #Uso size ya que me dice cuántas filas hay por categoría (es decir, cuántas ventas).
  #Uso resetindex para convertir Cantidad_vendida en una columna (evito que se quede como una serie aparte)

  resumen = (df_tienda.groupby('Categoria_producto').size().reset_index(name='Cantidad_vendida').sort_values(by='Cantidad_vendida', ascending=False))
  print(resumen)
  print("-" * 50)
```

```python
# Extra agregamos una columna que nos diga el importe vendido por categoria

for i, df_tienda in enumerate(lista_de_tiendas, start=1):
    print(f'Tienda {i} - Ventas por categoría')

    resumen_ventas = (df_tienda.groupby('Categoria_producto').agg(Cantidad_vendida=('Categoria_producto', 'size'),Importe_total=('Precio', 'sum')).reset_index().sort_values(by='Cantidad_vendida', ascending=False))
    resumen_ventas['Importe_total'] = resumen_ventas['Importe_total'].apply(lambda x: f"${x:,.2f}")
    print(resumen_ventas.to_string(index=False))
    print("-" * 50)
```

### **Nos percatamos que pese a que muebles es la categoria mas vendida, electrónicos y electrodomésticos son las que tienen un mayor importe de ventas** ###

```python

lista_de_tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df_tienda in enumerate(lista_de_tiendas, start=1):
    print(f"\n Tienda {i} - Productos más y menos vendidos")

    # Agrupar por producto y contar ventas
    ventas_por_producto = (df_tienda.groupby('Producto').size().reset_index(name='Cantidad_vendida').sort_values(by='Cantidad_vendida', ascending=False))

    # Producto más vendido
    top_producto = ventas_por_producto.iloc[0]
    # Producto menos vendido
    bottom_producto = ventas_por_producto.iloc[-1]

    print(f"Producto más vendido: {top_producto['Producto']} ({top_producto['Cantidad_vendida']} ventas)")
    print(f"Producto menos vendido: {bottom_producto['Producto']} ({bottom_producto['Cantidad_vendida']} ventas)")

    # Mostrar los 5 más vendidos y los 5 menos vendidos
    print("\nTop 5 productos más vendidos:")
    print(ventas_por_producto.head(5).to_string(index=False))

    print("\nTop 5 productos menos vendidos:")
    print(ventas_por_producto.tail(5).to_string(index=False))

    print("-" * 60)
```

### **Nos resultaria interesante ver si en realidad una mayor numero de ventas implica una mayor facturacion** ###

```python
lista_de_tiendas = [tienda, tienda2, tienda3, tienda4]

for i, df_tienda in enumerate(lista_de_tiendas, start=1):
    print(f"\n Tienda {i} - Productos más y menos vendidos")

    # Agrupar por producto: cantidad de ventas y monto generado
    ventas_por_producto = (df_tienda.groupby('Producto').agg(Cantidad_vendida=('Producto', 'size'),Importe_total=('Precio', 'sum')).reset_index().sort_values(by='Cantidad_vendida', ascending=False))

    # Formato de pesos
    ventas_por_producto['Importe_total'] = ventas_por_producto['Importe_total'].apply(lambda x: f"${x:,.2f}")

    # Producto más vendido
    top_producto = ventas_por_producto.iloc[0]
    # Producto menos vendido
    bottom_producto = ventas_por_producto.iloc[-1]

    print(f"\nProducto más vendido: {top_producto['Producto']} ({top_producto['Cantidad_vendida']} ventas, {top_producto['Importe_total']})")
    print(f"Producto menos vendido: {bottom_producto['Producto']} ({bottom_producto['Cantidad_vendida']} ventas, {bottom_producto['Importe_total']})")

    # Mostrar los 5 más vendidos y los 5 menos vendidos
    print("\nTop 5 productos más vendidos:")
    print(ventas_por_producto.head(5).to_string(index=False))

    print("\n\nTop 5 productos menos vendidos:")
    print(ventas_por_producto.tail(5).to_string(index=False))

    print("-" * 60)
```

### **Un mayor numero de ventas no implica un mayor importe de venta. Nos percatamos del enfoque en los productos que tiene cada tienda y su posible nicho.** ###

```python
import matplotlib.pyplot as plt
import pandas as pd

lista_de_tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = []
ingresos_totales = []

# --- 1. Recolección de Datos ---
for i, df_tienda in enumerate(lista_de_tiendas, start=1):

    nombre_tienda = f'Tienda {i}'

    # Calcular el ingreso total (suma de la columna 'Precio')
    ingreso_total = df_tienda['Precio'].sum()

    # Almacenar en listas
    nombres_tiendas.append(nombre_tienda)
    ingresos_totales.append(ingreso_total)

# --- 2. Creación del Gráfico de Barras ---
plt.figure(figsize=(10, 6))

# Crear el gráfico de barras
barras = plt.bar(nombres_tiendas, ingresos_totales, color=['dodgerblue', 'orange', 'green', 'red'])

# --- 3. Personalización del Gráfico ---
plt.title('Ingreso Total Generado por Cada Tienda', fontsize=16, fontweight='bold')
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Ingreso Total (en miles de millones)', fontsize=12)

# Formatear el eje Y para una mejor lectura (millones o miles de millones)
# Usaremos una función lambda para convertir los valores a formato legible
def format_billions(x, pos):
    'El argumento pos es necesario para el callback de formato de Matplotlib'
    if x >= 1e9:
        return f'{x*1e-9:.2f} MM'
    return f'{x*1e-6:.0f} M'

from matplotlib.ticker import FuncFormatter
plt.gca().yaxis.set_major_formatter(FuncFormatter(format_billions))

plt.grid(axis='y', linestyle='--', alpha=0.6)

# Añadir etiquetas de valor a cada barra para precisión
for bar in barras:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 10e6, f'${yval/1e6:,.0f} M', ha='center', va='bottom', rotation=45, fontsize=9)

plt.tight_layout()
plt.show()
```

##Visualizacion de envio promedio por tienda

```python
import matplotlib.pyplot as plt
import pandas as pd

lista_de_tiendas = [tienda, tienda2, tienda3, tienda4]
nombres_tiendas = []
costos_envio_promedio = []

# --- 1. Recolección de Datos ---
for i, df_tienda in enumerate(lista_de_tiendas, start=1):

    nombre_tienda_actual = f'Tienda {i}'

    # Calcular el costo de envío promedio
    # Asegúrate de que la columna 'Costo_envio' exista en tu DataFrame
    if 'Costo_envio' in df_tienda.columns:
        costo_envio_promedio = df_tienda['Costo_envio'].mean()
    else:
        # Manejo de error si la columna no existe
        print(f"Advertencia: La columna 'Costo_envio' no se encontró en {nombre_tienda_actual}. Se usará 0.")
        costo_envio_promedio = 0

    # Almacenar en listas
    nombres_tiendas.append(nombre_tienda_actual)
    costos_envio_promedio.append(costo_envio_promedio)

# --- 2. Creación del Gráfico de Barras ---
plt.figure(figsize=(10, 6))

# Crear el gráfico de barras
# Puedes usar los mismos colores si quieres consistencia con gráficos anteriores
colores_barras = ['dodgerblue', 'orange', 'green', 'red']
barras = plt.bar(nombres_tiendas, costos_envio_promedio, color=colores_barras)

# --- 3. Personalización del Gráfico ---
plt.title('Costo de Envío Promedio por Tienda', fontsize=16, fontweight='bold')
plt.xlabel('Tienda', fontsize=12)
plt.ylabel('Costo de Envío Promedio', fontsize=12)

# Añadir etiquetas de valor encima de cada barra
for bar in barras:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval, f'${yval:,.2f}',
             ha='center', va='bottom', fontsize=9, color='black')

plt.grid(axis='y', linestyle='--', alpha=0.6) # Solo cuadrícula horizontal para legibilidad
plt.ylim(0, max(costos_envio_promedio) * 1.1) # Ajustar límite superior del eje Y

plt.tight_layout()
plt.show()
```

##Mapa de calor por tienda

```python
import folium
from folium.plugins import HeatMap

for i, df_tienda in enumerate(lista_de_tiendas, start=1):
    resumen_ventas = (
        df_tienda
        .groupby(['lat', 'lon'])
        .agg(Importe_total=('Precio', 'sum'))
        .reset_index()
    )

    # Crear mapa centrado en la media de lat/lon
    mapa = folium.Map(
        location=[resumen_ventas['lat'].mean(), resumen_ventas['lon'].mean()],
        zoom_start=12
    )

    # Agregar HeatMap
    HeatMap(data=resumen_ventas[['lat', 'lon', 'Importe_total']].values.tolist(), radius=15).add_to(mapa)

    # Guardar el mapa en un archivo HTML
    mapa.save(f'mapa_tienda_{i}.html')
```

##Mapa de calor de todas las tiendas juntas

```python
import folium
from folium.plugins import HeatMap
import pandas as pd

# Lista de tus DataFrames de tiendas
lista_de_tiendas = [tienda, tienda2, tienda3, tienda4]

# Combinar todas las tiendas en un solo DataFrame
todos_los_datos = pd.concat(lista_de_tiendas)

# Eliminar filas con lat/lon vacíos
todos_los_datos = todos_los_datos.dropna(subset=['lat','lon'])

# Agrupar por ubicación y sumar el importe total
resumen_ventas = todos_los_datos.groupby(['lat','lon']).agg(Importe_total=('Precio','sum')).reset_index()

# Crear mapa centrado en la media de lat/lon
mapa = folium.Map(
    location=[resumen_ventas['lat'].mean(), resumen_ventas['lon'].mean()],
    zoom_start=5
)

# Crear HeatMap
HeatMap(
    data=resumen_ventas[['lat','lon','Importe_total']].values.tolist(),
    radius=15,       # tamaño de los puntos del heatmap
    max_zoom=15,
    blur=10
).add_to(mapa)

# Guardar mapa interactivo
mapa.save('mapa_calor_todas_las_tiendas.html')

display(mapa)
print("Mapa generado: mapa_calor_todas_las_tiendas.html")

```
